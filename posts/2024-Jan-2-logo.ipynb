```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# Linear classification plot
x = np.linspace(0, 2, 100)
y = 0.1 * x**2 + 2
plt.plot(x, y, label='Linear Classification', color='black', lw=3)

# Scattered points for two different classes
random_idx = np.random.randint(0, 100, 20)
class_1_points = y - 0.5*np.random.rand(100)
class_2_points = y + 0.5*np.random.rand(100)

# Scatter circles for Class 1 (hollow)
plt.scatter(x[random_idx], class_1_points[random_idx],
            edgecolors='black', facecolors='none', marker='o', label='Class 1', lw=2)

# Scatter diamonds for Class 2 (hollow)
plt.scatter(x[random_idx], class_2_points[random_idx], 
            edgecolors='black', facecolors='none', marker='D', label='Class 2', lw=2)

# Adding text in black
plt.text(0, 0, 'Sustainability Lab', fontsize=36, ha='center', color='black')
plt.text(0, -1, 'AI for Sustainability', fontsize=16, ha='center', color='black')

plt.axis('off')

plt.tight_layout()
```
